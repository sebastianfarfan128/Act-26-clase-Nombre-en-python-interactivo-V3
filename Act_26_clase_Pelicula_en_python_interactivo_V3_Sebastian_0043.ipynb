#### Clase Pelicula con 3 atributos funciones leerdatos(), mostrardatos()
#### Que sea interactivo ---> los inputs
#### crear 2 objetos pelicula1 y pelicula2
#### no usar while true, if, for, listas y diccionarios
[infoclasepelicula.png](infografiaclasepelicula.png)

```python
# clase Pelicula , atributos: titulo, genero, duracion
# funciones leerdatos(), mostrardatos()
# crear 2 objetos pelicula1, pelicula2

class Pelicula:
    def __init__(self, titulo, genero, duracion):
        self.titulo = titulo
        self.genero = genero
        self.duracion = duracion

    def leerdatos(self):
        self.titulo = input("Ingrese el título de la película: ")
        self.genero = input("Ingrese el género: ")
        self.duracion = int(input("Ingrese la duración (minutos): "))
        return self.titulo, self.genero, self.duracion

    def mostrardatos(self):
        print("Título: ", self.titulo)
        print("Género: ", self.genero)
        print("Duración: ", self.duracion, "min")
        return self.titulo, self.genero, self.duracion

# fin de la clase Pelicula
# creando los objetos pelicula1, pelicula2

print("🎬 Película Número 1")
pelicula1 = Pelicula("", "", 0)
pelicula1.leerdatos()
pelicula1.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*-*-*")

print("🎬 Película Número 2")
pelicula2 = Pelicula("", "", 0)
pelicula2.leerdatos()
pelicula2.mostrardatos()

print("-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("software desarrollado por Sebastian Farfan NL = 16")
```
